```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train_features=np.array(pd.read_csv('GRU_train.csv', index_col=0))
y_train_os=np.array(pd.read_csv('GRU_label.csv',index_col=0))
test_features=np.array(pd.read_csv('GRU_test.csv', index_col=0))
y_test=np.array(pd.read_csv('GRU_test_label.csv',index_col=0))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf1=GridSearchCV(SVC(gamma='auto',probability=True),{
    'C':[5,10,40,60,100],
    'kernel':['rbf','poly','linear']
    },cv=5,return_train_score=False)
clf1.fit(train_features,y_train_os)
result1=pd.DataFrame(clf1.cv_results_)
result1
```

```python
clf1.score(test_features,y_test)
```

```python
prediction_SF=clf1.predict(test_features)
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction_SF))
```

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(y_test,prediction_SF)
ax=plt.subplot()
sns.heatmap(cm1,annot=True)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels') 
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['B', 'M'])
ax.yaxis.set_ticklabels(['B', 'M']);
```

```python
from sklearn.metrics import roc_curve, auc

SF_fpr, SF_tpr, threshold = roc_curve(y_test, prediction_SF)
auc_SF = auc(SF_fpr, SF_tpr)


plt.figure(figsize=(5, 5), dpi=100)

plt.plot(SF_fpr, SF_tpr, marker='.', label='SVM (auc = %0.3f)' % auc_SF)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')

plt.legend()

plt.show()
```
